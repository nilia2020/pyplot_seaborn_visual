```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creo los datos aleatorios

data = np.random.randint(1,50,100)
```

### Histograma

```python
plt.hist(data, bins=10, rwidth=0.5)
%matplotlib inline
```

```python
plt.hist(data, bins=10, histtype='step')
%matplotlib inline
```

### Boxplot

```python
data = np.append(data, 100)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
%matplotlib inline
```

### Scatter

```python
N = 50
x = np.random.rand(N)
y= np.random.rand(N)
area = np.pi * (30 * x)**2
colors = y*x
```

```python
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
%matplotlib inline
```
