```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
# Asegurarse de que la columna 'sex' es categ√≥rica
tips['sex'] = tips['sex'].astype('category')
```

```python
sns.set_style('whitegrid')
```

```python
sns.countplot(data=tips, x='day',hue='sex')
%matplotlib inline
```

```python
plt.figure(figsize=(6, 6))
sns.stripplot(data=tips, x='day', hue='sex', y='total_bill', jitter=True, dodge=True, alpha=0.5, linewidth=1);
```

```python
plt.figure(figsize=(6, 6))
sns.swarmplot(data=tips, x='day', hue='sex', y='total_bill', dodge=True, alpha=0.5, linewidth=1);
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips,  y='total_bill', dodge=True,  linewidth=1);
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', hue='sex', y='total_bill', dodge=True,  linewidth=1);
sns.swarmplot(data=tips, x='day', hue='sex', y='total_bill', dodge=True,  linewidth=1, marker='o', palette='Set2');
```

```python
plt.figure(figsize=(6, 6))
sns.violinplot(data=tips, x='day', hue='sex', y='total_bill', dodge=True, split=True, palette='pastel');
```

```python
plt.figure(figsize=(6, 6))
sns.catplot(data=tips, x='day', hue='size', y='total_bill', kind='swarm', dodge=True, palette='autumn', col='time')
sns.catplot(data=tips, x='day', hue='sex', y='total_bill', kind='box', dodge=True, palette='autumn', col='time')
%matplotlib inline
```
