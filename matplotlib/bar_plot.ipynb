```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
from matplotlib import rcParams
import matplotlib.font_manager as fm#Set style
```

```python
# Data
countries = ["India", "Japan", "Mexico", "Colombia", "Germany"]
population = [1000, 800, 900, 1000, 300]
```

```python
# List available fonts
available_fonts = sorted([f.name for f in fm.fontManager.ttflist])
print(available_fonts)

print(plt.style.available)

plt.style.use('seaborn-v0_8-darkgrid')
rcParams['font.sans-serif'] = ['Ubuntu', 'DejaVu Sans',
                               'Source Code Pro', 'Lato']
plt.rcParams["font.sans-serif"] = ["Ubuntu"]
%matplotlib inline
```

```python
# Set bar_plot
plt.bar(
    countries, population, width=0.4, color=["#5c58b6", "#b457ce", "#5994ce", "#3a4e93"]
)
ax = plt.gca()
ax.set_xticks(range(len(countries)))

# Añadir texto en la parte superior derecha
ax.text(0.95, 0.95, 'Bar Chart', transform=ax.transAxes, fontsize=12, verticalalignment='top', horizontalalignment='right', math_fontfamily='cm')

# Define font properties
font_properties = FontProperties(family='Lato')
font_properties.set_stretch('ultra-expanded')

# Set x-tick labels with font properties
ax.set_xticklabels(countries, fontproperties=font_properties, size=11, rotation=45, ha='right')
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'))
           
%matplotlib inline
```

```python
plt.barh(
    countries, population, height=0.5, color=["#5c58b6", "#b457ce", "#5994ce", "#3a4e93"]
)
%matplotlib inline
```
