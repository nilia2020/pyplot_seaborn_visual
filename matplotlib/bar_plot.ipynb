```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
countries = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", "GERMANY"]
population = [1000, 800, 900, 1000, 300]
```

```python
#Set style
print(plt.style.available)
plt.style.use('seaborn-v0_8-darkgrid')

%matplotlib inline
```

```python
# Set bar_plot
plt.bar(
    countries, population, width=0.4, color=["#5c58b6", "#b457ce", "#5994ce", "#3a4e93"]
)
ax = plt.gca()
ax.set_xticks(range(len(countries)))
ax.set_xticklabels(countries, fontdict={'fontstretch':'expanded'})
%matplotlib inline
```
