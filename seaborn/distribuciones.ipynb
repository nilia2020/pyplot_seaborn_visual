```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
# Asegurarse de que la columna 'sex' es categ√≥rica
tips['sex'] = tips['sex'].astype('category')
```

```python
sns.set_style('whitegrid')
sns.histplot(data=tips, x='tip', kde=True, bins=30, hue='sex', stat='count', multiple='dodge')
%matplotlib inline
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True, bw_adjust=0.5)
%matplotlib inline
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count', )
%matplotlib inline
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='dodge')
%matplotlib inline
```
