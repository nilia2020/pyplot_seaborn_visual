```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
%matplotlib inline
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
%matplotlib inline
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
%matplotlib inline
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x, np.cos(x)/2, 'k')
axes[0,1].plot(x, np.cos(x)**3, 'r')
axes[0,2].plot(x, np.cos(x), "g")
axes[0,3].plot(x, np.cos(x)**2, 'c')
axes[1,0].plot(x, np.cos(x)+np.sin(x), 'y')
axes[1,1].plot(x, np.cos(x)+np.sin(x)**2, 'b')
axes[1,2].plot(x, np.cos(x)+np.sin(x)**3, 'r')
axes[1,3].plot(x, np.cos(x)+np.sin(x)**4, 'm')
%matplotlib inline
```

```python
fig, ((ax1, ax2, ax3, ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x)/2, 'k', label='1')
ax2.plot(x, np.cos(x)**3, 'r', label='2')
ax3.plot(x, np.cos(x), "g", label='3')
ax4.plot(x, np.cos(x)**2, 'c', label='4')
ax5.plot(x, np.cos(x)+np.sin(x), 'y', label='5')
ax6.plot(x, np.cos(x)+np.sin(x)**2, 'b', label='6')
ax7.plot(x, np.cos(x)+np.sin(x)**3, 'r', label='7')
ax8.plot(x, np.cos(x)+np.sin(x)**4, 'm', label='8')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
ax5.legend()
ax6.legend()
ax7.legend()
ax8.legend()
fig.tight_layout()
%matplotlib inline
```
