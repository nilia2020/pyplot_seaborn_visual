```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
# Asegurarse de que la columna 'sex' es categ√≥rica
tips['sex'] = tips['sex'].astype('category')
sns.set_style('whitegrid')
```

```python
markers = {'Lunch':'D', 'Dinner':'s'}
plt.figure(figsize=(10,8))
tips.head(2)
%matplotlib inline
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='kde', marginal_ticks=True, marginal_kws=dict(fill=False))
plt.show()
```
