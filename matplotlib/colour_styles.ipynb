```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,18)
```

```python
print(plt.style.available)
```

```python
plt.style.use('bmh')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x, x, 'r--')
ax.plot(x, x+1, 'b-')
ax.plot(x, x+2, 'gx')
ax.plot(x, x+3, 'y:')
%matplotlib inline
```

## Pyplot

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x, x, color='green', linewidth=5, linestyle='dashed', marker='x', markeredgecolor='yellow')
ax.plot(x, x+1, color='blue', linewidth=2, linestyle='-',marker='o', markerfacecolor='white')
ax.plot(x, x+2, color='#bbdd99', linewidth=7, linestyle='--', marker='v', markeredgecolor='green', markerfacecolor='red',markersize=10)
ax.plot(x, x+3, color='#ffccbb', alpha=0.6, linestyle=':', marker='P', markersize=8)
%matplotlib inline
```
