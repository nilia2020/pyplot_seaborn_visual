```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,18)
y = np.sin(x)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
ax1.plot(x,y, label='$sin(x)$')
ax1.set_title('Relación x - y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax2.plot(y,x, label='sin(y)')
ax2.set_title('Relación y - x')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()
%matplotlib inline
```

```python
plt.plot(x,y,label='función sen X ')
plt.title("Gráfico simple" )
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper right', bbox_to_anchor=(1, 0.9))
%matplotlib inline
```
