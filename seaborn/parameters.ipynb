```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips, x='total_bill')
```

```python
sns.set_theme(style='whitegrid', palette='pastel', color_codes=True)
sns.displot(tips, x='total_bill', y='tip')
%matplotlib inline
```

```python
sns.set_theme(style='whitegrid', palette='pastel', color_codes=True)
sns.displot(tips, x='total_bill', y='tip', hue='sex')
%matplotlib inline
```

```python
sns.set_theme(style='whitegrid', palette='pastel', color_codes=True)
sns.displot(tips, x='total_bill', hue='sex', kind='kde', legend=False, palette='Set1', alpha=0.5)
%matplotlib inline
```
